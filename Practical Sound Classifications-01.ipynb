```python
import os
import subprocess
import shutil
import pytube
```

```python
def download(vtype, link, num):
    yt = pytube.YouTube(link)  # 다운받을 동영상 URL 지정
    vids = yt.streams.all()

    # 영상 형식 리스트 확인
    # for i in range(len(vids)):
    #    print(i, '. ', vids[i])


    # 저장 경로 지정(Windows or mac)
    parent_dir = "./data5/" + str(vtype)

    #if os.path.exists(parent_dir):  # 이미 존재하는 경우, 삭제를 할까..
    #    try:
    #        shutil.rmtree(parent_dir)
    #    except OSError as e:
    #        if e.errno == 2:
    #            print  # 파일이나 디렉토리가 없음!
    #            'No such file or directory to remove'
    #            pass
    #        else:
    #            raise

    if os.path.exists(parent_dir):  # 이미 존재하는 경우, 삭제를 할까..
        pass
    else:
        # 디렉터리 생성
        os.mkdir(parent_dir)

    vnum = 1

    # 다운로드 수행
    vids[vnum].download(parent_dir)

    
    if vtype == 0:
        download.type = vtype
        download.num = 0
        
    else:
        if download.type != vtype:
            download.type = vtype
            download.num = 0
            
    # 파일 변환
    new_filename = str(download.num) + '.wav'
    default_filename = vids[vnum].default_filename

    # cmd 명령어 수행
    subprocess.call(
        ['ffmpeg', '-i', os.path.join(parent_dir, default_filename), os.path.join(parent_dir, new_filename)])

    print(num + '번 동영상 다운로드 및 mp3 변환 완료!')
    
    test = os.listdir(parent_dir)
    
    for item in test:
        if item.endswith(".webm"):
            os.remove(os.path.join(parent_dir, item))

    download.num = download.num + 1

    
```

```python
download.num = -1
download.type = -1

f = open("./download_list5.txt")
links = f.readlines()
f.close()

data_path = "./data5"

if os.path.exists(data_path):
    pass
else:
    os.mkdir(data_path)


for i in range(len(links)):
 
    if i % 2 == 0:
        vtype = int(links[i])
        print(links[i+1])
        download(vtype, links[i+1], str(int((i + 1)/2)+1))
    
```
